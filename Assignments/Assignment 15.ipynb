```python
from ip2geotools.databases.noncommercial import DbIpCity
def getcor(x):
    response = DbIpCity.get(x['IP'], api_key='free')
    return pd.Series((response.latitude,response.longitude))

df2[['latitude','longitude']] = df2.apply(getcor, axis=1)
```

```python
df2
```

```python
from geopy.distance import great_circle

def get_dist(x):
    try:
        loc_1 = (x['latitude'], x['longitude'])
        loc_2 = (30.13986360774072, 31.624661271290293)
        return great_circle(loc_1, loc_2).kilometers
    except:
        return 'NaN'
df2['Dist_From_Egypt'] = df2.apply(get_dist, axis=1)
df2
```
