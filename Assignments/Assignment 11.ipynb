# GPU system and some examples

```python
def Grades(posttest):
    if posttest > 50 and posttest <= 60:
        return "D"
    if posttest > 60 and posttest <= 75:
        return "C"
    if posttest > 75 and posttest <= 85:
        return "B"
    if posttest > 85 and posttest <= 100:
        return "A"
    else:
        return "F"
    
df['posttest'].apply(Grades)
```

```python
sum(df['posttest'].apply(Grades) == "A")
```

```python
sum(df['posttest'].apply(Grades) == "F")
```

```python
df[(df['posttest'].apply(Grades) == "F") & (df['school_type'] == "Public")].count()
```

# How many one got posttest results = "A" and from public school and male ?

```python
sum(df[(df['posttest'].apply(Grades) == "A") & (df['school_type'] == "Public") & (df['gender'] == "Male")]["student_id"].value_counts())
```

# How many one from "CCAAW" School got an "A" ?

# What are the ID's of the students who got an "A" whether at pretest or posttest from "GOOBU" School ?

```python
df[(df['school'] == 'GOOBU') & (df['posttest'].apply(Grades) == "A" ) | (df['school'] == 'GOOBU') & (df['pretest'].apply(Grades) == "A" )]["student_id"]
#Comment on results: the results indicates that no one from GOOBU School got an "A" whether at pretest or posttest, indicates that the school is فشله
```

# The mean of female students in public schools with standard teaching method posttest marks

```python
X = df[(df['gender'] == 'Female') & (df['school_type'] == 'Public') & (df['teaching_method'] == 'Standard')]['posttest'].mean()
averageGPU = Grades(X)
print(X)
print(averageGPU)
```
