# gapminder-FiveYearData

```python
gm = pd.read_csv('gapminder-FiveYearData.csv')
gm
```

```python
gm.info()
```

```python
gm.describe()
```

```python
gm_table = pd.pivot_table(data = gm, values = 'pop', index = 'continent', columns ='year')
gm_table
```

```python
a4_dims = (12, 9)
fig, ax = plt.subplots(figsize=a4_dims)
sns.heatmap(gm_table, cmap = 'inferno', annot = True, ax = ax, linewidths=1,
    linecolor='white')
```
