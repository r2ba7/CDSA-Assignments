```python
import sys
import BankAssignment
       
def namecheck(name):
    if name == "0":
        sys.exit()
    for char in name:
        if char.isalpha():
            return True
        else:
            print("Invalid name")
            return False

def gendercheck(gender):
    if gender == "0":
        sys.exit()
    if gender == "Male" or gender == "Female" or gender == "male" or gender == "female" or gender == "Other" or gender == "other":
        return True
    else:
        print("Wrong gender, restarting program")
        main()

def agecheck():
    try:
        age = int(input("Please enter your age: "))
    except:
        print("Wrong age format")
        main()
    else:
        if(age == 0):
            sys.exit()
        elif(age > 0 and age < 18):
            print("Younger than 18")
            agecheck()
        else:
            return age

#def balancecheck():
 #   try:
 #       balance = int(input("Please the balance: "))
  #  except:
  #      print("Invalid balance")
   #     main()
   # else:
    #    return balance
        
def main():
    while True:
        print("To exit type 0")
        name = str(input("Please enter your name: "))
        Flag = namecheck(name)
        if Flag == False:
            main()
        age = agecheck()
        gender = str(input("Please enter your gender: "))
        if gendercheck(gender) == True:
            print("account created successfuly!")
            User1 = BankAssignment.User(name, age, gender)
            ID = input("Please enter your ID: ") 
            Password = input("Please choose your Password: ")
            #balance = balancecheck()
            User1_Bank = BankAssignment.Bank(User1)
            User1_Bank.setData(ID, Password)
            order = int(input("What to do ?"))
            print("0 = exit, 1 = show details, 2 = view balance, 3 = withdraw, 4 = deposit, 5 = sign out")
            if order == 0:
                sys.exit()
            elif order == 1:
                 User1.showdetails()
            elif order == 2:
                User1_Bank.viewbalance()
            elif order == 3:
                ##Error Handling
                amount = int(input("Choose your amount: "))
                User1_Bank.withdraw(amount)
            elif order == 4:
                amount = int(input("Choose your amount: "))
                User1_Bank.deposit(amount)
            elif order == 5:
                main()
            else:
                print("Wrong input")
                continue 
        else:
            print("Wrong Input")
            break
            
if __name__ == "__main__":
    main()
```
