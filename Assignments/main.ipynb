```python
import sys
import BankAssignment as BN
       
def namecheck(name):
    if name == "0":
        sys.exit()
    for char in name:
        if char.isalpha() or char == " ":
            continue
        else:
            print("Invalid name")
            return False

def gendercheck(gender):
    if gender == "0":
        sys.exit()
    if gender == "Male" or gender == "Female" or gender == "male" or gender == "female" or gender == "Other" or gender == "other":
        return True
    else:
        print("Wrong gender, restarting program")
        main()

def agecheck():
    try:
        age = int(input("Please enter your age: "))
    except:
        print("Invalid age format")
        agecheck()
    else:
        if(age == 0):
            sys.exit()
        elif(age > 0 and age < 18):
            print("Younger than 18")
            agecheck()
        else:
            return age

def amountcheck():
    try:
        x = int(input("Enter Amount: "))
    except:
        print("Invalid Operation")
        amountcheck()
    else:
        if(x == 0):
            sys.exit()
        elif(x < 0):
            print("Invalid amount")
            amountcheck()
        else:
            return x
        
def Operationcheck():
    try:
        x = int(input("Enter Operation Number: "))
    except:
        print("Invalid Operation")
    else:
        if(x == 0):
            sys.exit()
        else:
            return x

        
def main():
    while True:
        print("To exit type 0")
        name = str(input("Please enter your name: "))
        Flag = namecheck(name)
        if Flag == False:
            main()
        age = agecheck()
        gender = str(input("Please enter your gender: "))
        if gendercheck(gender) == True:
            print("account created successfuly!")
            User1 = BN.User(name, age, gender)
            ID = input("Please enter your ID: ") 
            Password = input("Please choose your Password: ")
            User1Bank = BN.Bank(name, age, gender)
            User1Bank.set_Data(ID, Password)
            while True:
                print("0 = exit, 1 = show details, 2 = view balance, 3 = withdraw, 4 = deposit, 5 = sign out")
                print("Operation: ")
                order = Operationcheck()
                if order == 0:
                    sys.exit()
                elif order == 1:
                     User1Bank.showdetails()
                elif order == 2:
                    User1Bank.viewbalance()
                elif order == 3:
                    amount = amountcheck()
                    User1Bank.withdraw(amount)
                elif order == 4:
                    amount = amountcheck()
                    User1Bank.deposit(amount)
                elif order == 5:
                    main()
                else:
                    print("Wrong input")
                    continue 
        else:
            print("Wrong Input")
            break
            
if __name__ == "__main__":
    main()
```
