# 911
### Visualize the 911 dataset using those features twp (the town), Reason, hour, month, day of week, Year

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(rc={'figure.figsize':[8,8]}, font_scale = 1)
```

```python
df1 = pd.read_csv('911.csv')
df1
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1.head()
```

```python
plt.subplot(2,1,1)
sns.countplot(data = df1, x = 'Year', palette = 'twilight')
plt.subplot(2,1,2)
sns.countplot(data = df1, x = 'Reason', palette = 'mako')
```

```python
df1['twp'].unique()
```

```python
plt.subplot(2,1,1)
sns.countplot(data = df1, x = df1['twp'], order = df1['twp'].value_counts().index[0:5], palette = 'mako')
plt.subplot(2,1,2)
sns.countplot(data = df1, x = df1['twp'], order = df1['twp'].value_counts().index[0:5], hue = 'Reason', palette = 'plasma')
```

```python
plt.subplot(2,1,1)
sns.countplot(data = df1, x = df1['hour'])
plt.subplot(2,1,2)
sns.countplot(data = df1, x = df1['hour'], hue = 'Reason')
```

```python
plt.subplot(2,1,1)
sns.countplot(data = df1, x = df1['month'])
plt.subplot(2,1,2)
sns.countplot(data = df1, x = df1['month'], hue = 'Reason')
```

```python
df1.info()
```

```python
sns.catplot(x="Reason", kind="count", palette="ch:.25", data=df1, hue = 'month')
```

```python
sns.catplot(x="day of week", kind="count", palette="ch:.25", data=df1, hue = 'Year')
```

```python
sns.catplot(x="month", kind="count", palette="mako", data=df1, hue = 'Year')
```

```python
df1_table = pd.pivot_table(data = df1, values = 'hour', index = df1['twp'], columns ='Reason', aggfunc= np.sum)
df1_table
```

# Titanic

### Visualize the titanic dataset using those features survived, sex, age, fare, class, who, embark_town, alive, alone

```python
df2 = pd.read_csv('titanic.csv')
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2.head()
```

```python
fig, axes = plt.subplots(3, figsize = (10,10))
fig.suptitle('Plots')
sns.countplot( data = df2, x = 'survived', hue = 'sex', ax = axes[0])
sns.countplot( data = df2, x = 'survived', hue = 'who', palette = 'Blues_r', ax = axes[1])
sns.countplot( data = df2, x = 'survived', hue = 'alone', palette = 'ocean', ax = axes[2])
```

```python
sns.histplot(data = df2, x = 'fare', bins = 15, color = 'k')
```

```python
sns.boxplot(x='who', y ='fare', data= df2 , hue= 'survived')
```

```python
sns.boxplot(x='sex', y ='age', data= df2 , hue= 'survived', palette = 'gray')
```

```python
sns.stripplot(x='sex', y ='age', data= df2 , hue= 'survived', palette = 'mako')
```

```python
sns.stripplot(x='who', y ='age', data= df2 , hue= 'survived', palette = 'mako')
```

```python
sns.boxplot(x='survived', y ='age', data= df2 , hue= 'who', palette = 'mako')
```

```python
sns.violinplot(x= 'class', y='age' , data=df2, hue= 'sex', split= True )
```

```python
fig, axes = plt.subplots(2,2 , figsize= (20, 10)) 
fig.suptitle ('Plots')
sns.boxplot(ax=axes [0][0], x = 'embark_town', y ='age' , data= df2 , hue= 'alive', palette = 'twilight', dodge = True)
sns.boxplot(ax=axes [0] [1], x = 'alive', y='age', data=df2, hue= 'alone', palette = 'gist_stern')
sns.stripplot (ax = axes [1][0], x = 'class', y ='age', data= df2, hue = 'sex')
sns.violinplot(ax = axes [1][1], x = 'survived', y='age', data=df2, hue= 'alone' , palette='mako')
```

```python
sns.barplot(x='who', y ='age', data = df2, estimator=np.sum, hue='survived' , palette='mako')
```

```python
sns.jointplot(x = 'age' , y ='fare' , data= df2)
```

```python
sns.pairplot(df2)
```

```python
sns.barplot(x='who' , y ='age' , data=df2 , palette='Accent', hue='embark_town')
```
