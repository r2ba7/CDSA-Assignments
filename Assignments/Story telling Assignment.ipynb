### Plotting Confirmed/Deaths/Recovered/Active for a sample of 10 nations

```python
plt.subplot(5,1,1)
sns.countplot(data = df, x = 'WHO Region', palette = 'Blues', order = df['WHO Region'].value_counts().index)
plt.subplot(5,1,2)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'Confirmed', palette = 'mako', hue = 'WHO Region')
plt.subplot(5,1,3)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'Deaths', palette = 'mako', hue = 'WHO Region')
plt.subplot(5,1,4)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'Recovered', palette = 'mako', hue = 'WHO Region')
plt.subplot(5,1,5)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'Active', palette = 'mako', hue = 'WHO Region')
```

### Plotting New cases/New deaths/New recovered for a sample of 10 nations

```python
plt.subplot(3,1,1)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'New cases', palette = 'mako', hue = 'WHO Region')
plt.subplot(3,1,2)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'New deaths', palette = 'mako', hue = 'WHO Region')
plt.subplot(3,1,3)
sns.barplot(data = df, x = df['Country/Region'].sample(n = 10), y = 'New recovered', palette = 'mako', hue = 'WHO Region')
```

### Plotting Confirmed/Recovered/Deaths with hue as continent to know which continents were vulnerable
